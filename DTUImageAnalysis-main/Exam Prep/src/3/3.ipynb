```python
def histogram_stretch(img_in):    
    img_float = img_as_float(img_in)
    min_val = img_float.min()
    max_val = img_float.max()
    min_desired = 0.0
    max_desired = 1.0
    
    # Implementation
    img_out = (max_desired - min_desired)/(max_val-min_val)*(img_float-min_val)

    # img_as_ubyte will multiply all pixel values with 255.0 before converting to unsigned byte
    return img_as_ubyte(img_out)
```

```python
im_strech = histogram_stretch(im_org)
plt.hist(im_strech.ravel(), bins=256)
plt.title('Image histogram')
io.show()
```

```python
def gamma_map(img_in, gamma):    
    img_float = img_as_float(img_in)
    
    # Implementation
    img_out = np.power(img_float, gamma)

    # img_as_ubyte will multiply all pixel values with 255.0 before converting to unsigned byte
    return img_as_ubyte(img_out)
```

```python
gamma_half = gamma_map(im_org, 0.5)
io.imshow(gamma_half)
io.show()
```

```python
gamma_two = gamma_map(im_org, 2)
io.imshow(gamma_two)
io.show()
```

```python
gamma_five = gamma_map(im_org, 5)
io.imshow(gamma_five)
io.show()
```

```python
def threshold_image(img_in, thres):
    img_float = img_as_float(img_in)
    img_out = img_float > thres
    
    return img_as_ubyte(img_out)
```

```python
thresholdTest = threshold_image(im_org, 0.6)
io.imshow(thresholdTest)
io.show()
```

```python
def detect_dtu_signs(img_in):
    r_comp = img_in[:, :, 0]
    g_comp = img_in[:, :, 1]
    b_comp = img_in[:, :, 2]
    segm_blue = (r_comp < 10) & (g_comp > 85) & (g_comp < 105) & (b_comp > 180) & (b_comp < 200)
    return segm_blue 
```

```python

im3_detected = detect_dtu_signs(im3)
io.imshow(im3_detected)
io.show()
```

```python
def detect_dtu_signs2(img_in):
    r_comp = img_in[:, :, 0]
    g_comp = img_in[:, :, 1]
    b_comp = img_in[:, :, 2]
    segm_blue = (r_comp < 10) & (g_comp > 85) & (g_comp < 105) & (b_comp > 180) & (b_comp < 200)
    segm_red = (r_comp > 150) & (g_comp < 75) & (b_comp < 75)
    return (segm_blue | segm_red)
```

```python
im3_detected2 = detect_dtu_signs2(im3)
io.imshow(im3_detected2)
io.show()
```
