```python
from scipy.ndimage import correlate
from skimage import color, io, measure, img_as_ubyte
from skimage.measure import profile_line
from skimage.transform import rescale, resize
import matplotlib.pyplot as plt
import numpy as np
from skimage.util import img_as_float
from skimage.util import img_as_ubyte
from skimage.util import img_as_int
from skimage.filters import threshold_otsu
from skimage.filters import median
from skimage.filters import gaussian
from skimage.filters import prewitt_h
from skimage.filters import prewitt_v
from skimage.filters import prewitt
from skimage.morphology import erosion, dilation, opening, closing
from skimage.morphology import disk 
```

```python
def plot_comparison(original, filtered, filter_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')
    io.show()
```

```python
im_org =io.imread("data/legO_5.png")
im_gray =  color.rgb2gray(im_org)
thresh = threshold_otsu(im_gray)
bin_img = im_gray >= thresh
plot_comparison(im_org, bin_img, 'Binary image')
```

```python
footprint = disk(2)
# Check the size and shape of the structuring element
print(footprint)
```

```python
eroded = erosion(bin_img, footprint)
plot_comparison(bin_img, eroded, 'erosion')
```

```python
dilated = dilation(bin_img, footprint)
plot_comparison(bin_img, dilated, 'dilation')
```

```python
opened = opening(bin_img, footprint)
plot_comparison(bin_img, opened, 'opening')
```

```python
closed = closing(bin_img, footprint)
plot_comparison(bin_img, closed, 'closing')
```

```python
def compute_outline(bin_img):
    """
    Computes the outline of a binary image
    """
    footprint = disk(1)
    dilated = dilation(bin_img, footprint)
    outline = np.logical_xor(dilated, bin_img)
    return outline
```

```python
im_outline = compute_outline(bin_img)
plot_comparison(bin_img, im_outline, 'outline')
```

```python
open_close_img = closing(opening(bin_img, disk(15)), disk(1))
open_close_outline = compute_outline(open_close_img)
plot_comparison(bin_img, open_close_outline, 'opening')
```

```python
im_2 = color.rgb2gray(io.imread("data/lego_3.png"))
thresh_2 = threshold_otsu(im_2)
bin_img_2 = im_2 >= thresh_2
plot_comparison(im_2, bin_img_2, 'Binary image')
```

```python
outline_2 = compute_outline(bin_img_2)
plot_comparison(bin_img_2, outline_2, 'Outline image')
```

```python
closing_2 = closing(closing(outline_2, disk(25)), disk(10))
plot_comparison(bin_img_2, closing_2, 'Outline image')
```

```python
im = color.rgb2gray(io.imread("data/lego_9.png"))
t = threshold_otsu(im)
binary = im <= t
plot_comparison(im, binary, 'Binary image')
```

```python
ol = compute_outline(binary)
plot_comparison(binary, ol, 'Outline')
```

```python
im_closed = closing(binary, disk(3))
ol_closed = compute_outline(im_closed)
plot_comparison(binary, ol_closed, 'closed')
```

```python
im_eroded = erosion(im_closed, disk(45))
plot_comparison(binary, im_eroded, 'eroded')
```

```python
im_dialted = dilation(im_eroded, disk(25))
plot_comparison(binary, im_dialted, 'dialted')
```
