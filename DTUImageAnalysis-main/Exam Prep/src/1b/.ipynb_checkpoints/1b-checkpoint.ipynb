```python
from skimage import color, io, measure, img_as_ubyte
from skimage.measure import profile_line
from skimage.transform import rescale, resize
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
in_dir = "data/"
# X-ray image
txt_name = "irisdata.txt"
iris_data = np.loadtxt(in_dir + txt_name, comments="%")
# x is a matrix with 50 rows and 4 columns
x = iris_data[0:50, 0:4]
```

```python
n_feat = x.shape[1]
n_obs = x.shape[0]
print(f"Number of features: {n_feat} and number of observations: {n_obs}")
```

```python
sep_l = x[:, 0]
sep_w = x[:, 1]
pet_l = x[:, 2]
pet_w = x[:, 3]
```

```python
# Use ddof = 1 to make an unbiased estimate
var_sep_l = sep_l.var(ddof=1)
var_sep_w = sep_w.var(ddof=1)
var_pet_l = pet_l.var(ddof=1)
var_pet_w = pet_w.var(ddof=1)
print(f"Sepal length variance: {var_sep_l}\nSepal width variance: {var_sep_w}\nPetal length variance: {var_pet_l}\nPetal width variance: {var_pet_w}\n")
```

```python
covariance_sep_l_pet_l = 1/49*(sep_l*pet_l).sum()
covariance_sep_w_sep_l = 1/49*(sep_l*sep_w).sum()
print(f"Septal length and widh: {covariance_sep_w_sep_l} | Septal length and petal length: {covariance_sep_l_pet_l}")
```

```python
plt.figure() # Added this to make sure that the figure appear
# Transform the data into a Pandas dataframe
d = pd.DataFrame(x, columns=["Sepal length", "Sepal width","Petal length", "Petal width"])
sns.pairplot(d)
plt.show()
```

```python
mn = np.mean(x, axis=0)
data = x - mn
```

```python
C_x = 1/49*(np.matmul(x,np.transpose(x)))
```

```python
values, vectors = np.linalg.eig(C_x)
```

```python
v_norm = values / values.sum() * 100
plt.plot(v_norm)
plt.xlabel("Principal component")
plt.ylabel("Percent explained variance")
plt.ylim([0, 100])
plt.xlim([0,4])
plt.show()
print(f"First pc {values[0]}")
```

```python
pc_proj = vectors.T.dot(vectors.T)
```

```python
from sklearn import decomposition
```

```python
pca = decomposition.PCA()
pca.fit(x)
values_pca = pca.explained_variance_
exp_var_ratio = pca.explained_variance_ratio_
vectors_pca = pca.components_
```

```python
data_transform = pca.transform(data)
```
